```
import numpy as np
import os
import pandas as pd
```

```
features_csv=["./features_"+str(x)+".csv" for x in range(1,37)]
empty = ['./features_27.csv','./features_32.csv','./features_33.csv']
features_csv = [ele for ele in features_csv if ele not in empty]
features_csv
```

```
labels_csv=["./labels_"+str(x)+".csv" for x in range(1,37)]
empty = ['./labels_27.csv','./labels_32.csv','./labels_33.csv']
labels_csv = [ele for ele in labels_csv if ele not in empty]
labels_csv
```

```
features=[]
labels=[]
for x in features_csv:
  features.append(pd.read_csv(x,header=None))

for x in labels_csv:
  labels.append(pd.read_csv(x,header=None))
```

```
X=pd.concat(features)
X.shape
```

```
y=pd.concat(labels)
y.shape
```

```
y.replace(10,1,True)
y_np=y.values
y_np.reshape(-1,1)
y_np.shape
X_np=X.values
```

```
from numpy import inf
X_np[X_np==inf] = 1
```

```
from sklearn.preprocessing import scale
X_n = scale(X_np)
X_n
```

```
X_train = X_n[:round(len(X_n)*0.80)]
X_test = X_n[round(len(X_n)*0.80):]
y_train = y_np[:round(len(X_np)*0.80)]
y_test = y_np[round(len(X_np)*0.80):]
```

```
X_train = np.expand_dims(X_train, axis=2)
X_test = np.expand_dims(X_test, axis=2)
```

```
from keras.models import Sequential
from keras.layers import Dense, Flatten
from keras.layers.convolutional import Conv1D
from keras.optimizers import Adam, RMSprop
from keras.layers import Dropout
## Create Model ##

model = Sequential()

model.add(Conv1D(32, kernel_size = 3, activation = 'relu', input_shape = (4,1)))
model.add(Flatten())
model.add(Dense(16, activation='relu'))
model.add(Dense(8, activation = 'relu'))
model.add(Dropout(0.2))
model.add(Dense(1,activation = 'sigmoid'))

optimizer = Adam(lr=0.00001)
model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])

model.fit(X_train, y_train, validation_data = (X_test,y_test), epochs = 100)
```

```
from sklearn.metrics import roc_curve, roc_auc_score, f1_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn import metrics
y_pred = model.predict_classes(X_test)
y_score = model.predict_proba(X_test)
acc = accuracy_score(y_test, np.array(y_pred))
f1_score = metrics.f1_score(y_test, y_pred)
roc = metrics.roc_auc_score(y_test, y_score)
print([acc,f1_score,roc])     
print(confusion_matrix(y_test, y_pred))
```
