```
def capture_frame(sec):
  start_time=180000          # frames are extracted starting from 3min(180000 ms) till the end of video
  vid.set(cv2.CAP_PROP_POS_MSEC,start_time+sec*1000)
  res, frame=vid.read()
  return res,frame
```

```
# This block of code extracts  1fps from videos of each 25 participants
for i in [25]:
  frame_data = []
  for j in [0,10]:
    path1 = './'+str(i)+'/'+str(j)+'.mp4'
    path2 = './'+str(i)+'/'+str(j)+'.mov'
    path3 = './'+str(i)+'/'+str(j)+'.MOV'
    path4 = './'+str(i)+'/'+str(j)+'.MP4'
    if  os.path.isfile(path1):
       vid=cv2.VideoCapture('./'+str(i)+'/'+str(j)+'.mp4')
    elif  os.path.isfile(path2):
       vid=cv2.VideoCapture('./'+str(i)+'/'+str(j)+'.mov')
    elif os.path.isfile(path3):
       vid=cv2.VideoCapture('./'+str(i)+'/'+str(j)+'.MOV')
    elif os.path.isfile(path4):
       vid=cv2.VideoCapture('./'+str(i)+'/'+str(j)+'.MP4')
    else :
       vid=cv2.VideoCapture('./'+str(i)+'/'+str(j)+'.m4v')
    time=0
    res,frame= capture_frame(time)
    time+=1
    print(i,j)
    while(res):
        label = [j/10]
        frame_data.append([np.array(frame),np.array(label)])
        res,frame=capture_frame(time)
        time+=1

    vid.release()
    cv2.destroyAllWindows()

  frame_data=np.array(frame_data)
  np.save(open('./frames/ID_'+str(i)+'.npy','wb'),frame_data)
```
