```
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```
k = 25
frame = np.load('./ID_'+str(k)+'.npy',allow_pickle= True)
```

```
frame.shape
```

```
frame[0][0].shape
```

```
img = frame[449][0]
img = np.rot90(img)
img = np.rot90(img)
img = np.rot90(img)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = cv2.resize(img,(227,227))
```

```
plt.imshow(img)
```

```
for i in range(449,frame.shape[0]):
  img = frame[i][0]
  img = np.rot90(img)
  img = np.rot90(img)
  img = np.rot90(img)
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  img = cv2.resize(img,(227,227))
  frame[i][0] = img
```
